```python
columns_to_plot = ['Country', 'CountryIncomeLevel2021', 'Gender',
                   'Education', 'IncomeFeelings', 'INCOME_5', 'EMP_2010', 'Urbanicity',
                   'HouseholdSize', 'ChildrenInHousehold', 'Q1', 'Q6',  'Q8', 'Q9',
                   'Q10Q11Recode']

num_plots = len(columns_to_plot)
rows = num_plots // 2 if num_plots % 2 == 0 else num_plots // 2 + 1
cols = 2 if num_plots > 1 else 1

# Create a subplot
fig, axes = plt.subplots(rows, cols, figsize=(12, 30),dpi = 200)

# Flatten the axes array if necessary
if num_plots > 1:
    axes = axes.flatten()

# Create count plots for each specified column
for i, col in enumerate(columns_to_plot):
    ax = axes[i]
    sns.countplot(data=df, x=col, ax=ax)
    ax.set_title(f'Countplot of {col}')

# Remove any empty subplots (if num_plots is odd)
if num_plots % 2 != 0:
    fig.delaxes(axes[-1])

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
